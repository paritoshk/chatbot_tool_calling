```python
db = FakeDatabase()

def process_tool_call(tool_name, tool_args):
    if tool_name == "get_user":
        return db.get_user(tool_args["key"], tool_args["value"])
    elif tool_name == "get_order_by_id":
        return db.get_order_by_id(tool_args["order_id"])
    elif tool_name == "get_customer_orders":
        return db.get_customer_orders(tool_args["customer_id"])
    elif tool_name == "cancel_order":
        return db.cancel_order(tool_args["order_id"])
    else:
        return f"Unknown tool: {tool_name}"
```

```python
def extract_reply(text):
    pattern = r'<reply>(.*?)</reply>'
    match = re.search(pattern, text, re.DOTALL)
    if match:
        return match.group(1).strip()
    else:
        return text
```

```python

MODEL = "Qwen/Qwen3-235B-A22B-fp8-tput"

# A simple test message to see if the API works
test_message = "I need to check my order status. My email is john@gmail.com"

try:
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {
                "role": "system", 
                "content": "You are a customer support agent for TechNova. Use the tools available to help customers."
            },
            {
                "role": "user", 
                "content": test_message
            }
        ],
        tools=tools,
        temperature=0.2,
        max_tokens=1024
    )
    
    print("API Response:")
    print(json.dumps(response.model_dump(), indent=2))
    
    # Check if the model used a tool
    if hasattr(response.choices[0].message, 'tool_calls') and response.choices[0].message.tool_calls:
        print("\nTool calls detected!")
        for tool_call in response.choices[0].message.tool_calls:
            print(f"Tool: {tool_call.function.name}")
            print(f"Arguments: {tool_call.function.arguments}")
    else:
        print("\nNo tool calls in this response")
        
except Exception as e:
    print(f"API Error: {e}")
```

```python
def debug_message_flow():
    """Shows how messages flow through the conversation with tool calls"""
    client = Together(api_key=os.getenv("TOGETHER_API_KEY"))
    MODEL = "Qwen/Qwen3-235B-A22B-fp8-tput"
    
    system_prompt = "You are a TechNova customer support bot. Use tools to help customers."
    
    # Initial conversation
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": "I need to cancel my order but I don't know the order ID. My email is john@gmail.com"}
    ]
    
    print("STEP 1: Initial user query")
    print(f"User: {messages[-1]['content']}")
    
    # First call - model should use get_user tool
    print("\nSTEP 2: First model response (should use get_user tool)")
    response1 = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        tools=tools,
        temperature=0.2,
        max_tokens=1024
    )
    
    if not (hasattr(response1.choices[0].message, 'tool_calls') and response1.choices[0].message.tool_calls):
        print("Error: Model did not use tools as expected")
        return
        
    tool_call1 = response1.choices[0].message.tool_calls[0]
    print(f"Model wants to use: {tool_call1.function.name}")
    print(f"Arguments: {tool_call1.function.arguments}")
    
    # Add assistant response with tool call to messages
    messages.append({
        "role": "assistant",
        "content": response1.choices[0].message.content,
        "tool_calls": response1.choices[0].message.tool_calls
    })
    
    # Process tool call and add result
    args1 = json.loads(tool_call1.function.arguments)
    result1 = process_tool_call(tool_call1.function.name, args1)
    
    print("\nSTEP 3: Tool result")
    print(json.dumps(result1, indent=2))
    
    # Add tool result to messages
    messages.append({
        "role": "tool",
        "tool_call_id": tool_call1.id,
        "name": tool_call1.function.name,
        "content": json.dumps(result1)
    })
    
    # Second call - model should use get_customer_orders tool
    print("\nSTEP 4: Second model response (should use get_customer_orders tool)")
    response2 = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        tools=tools,
        temperature=0.2,
        max_tokens=1024
    )
    
    if not (hasattr(response2.choices[0].message, 'tool_calls') and response2.choices[0].message.tool_calls):
        print("Model response without tool call:")
        print(response2.choices[0].message.content)
        return
        
    tool_call2 = response2.choices[0].message.tool_calls[0]
    print(f"Model wants to use: {tool_call2.function.name}")
    print(f"Arguments: {tool_call2.function.arguments}")
    
    # Add assistant response with tool call to messages
    messages.append({
        "role": "assistant",
        "content": response2.choices[0].message.content,
        "tool_calls": response2.choices[0].message.tool_calls
    })
    
    # Process tool call and add result
    args2 = json.loads(tool_call2.function.arguments)
    result2 = process_tool_call(tool_call2.function.name, args2)
    
    print("\nSTEP 5: Second tool result (customer orders)")
    print(json.dumps(result2, indent=2))
    
    # Add tool result to messages
    messages.append({
        "role": "tool",
        "tool_call_id": tool_call2.id,
        "name": tool_call2.function.name,
        "content": json.dumps(result2)
    })
    
    # Final response to user
    print("\nSTEP 6: Final model response to user")
    final_response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        temperature=0.2,
        max_tokens=1024
    )
    
    print(f"TechNova Support: {final_response.choices[0].message.content}")

# Uncomment to run the debug flow
# debug_message_flow()
```

```python
from IPython.display import display, clear_output
def simple_jupyter_chat():
    # Initialize Together client
    client = Together(api_key=TOGETHERAI_API_KEY)
    MODEL = "Qwen/Qwen3-235B-A22B-fp8-tput"
    
    system_prompt = """
    You are a customer support chat bot for an online retailer called TechNova.
    Your job is to help users look up their account, orders, and cancel orders.
    Be helpful and brief in your responses.
    You have access to a set of tools, but only use them when needed.
    If you do not have enough information to use a tool correctly, ask a user follow up questions to get the required inputs.
    Do not call any of the tools unless you have the required data from a user.
    """
    
    # Initialize conversation
    messages = [
        {"role": "system", "content": system_prompt}
    ]
    
    print("TechNova Customer Support Bot (type 'exit' to end)")
    print("="*50)
    
    # This will be called for each user message
    def process_message(user_input):
        # Add user message to conversation
        messages.append({"role": "user", "content": user_input})
        
        # Process with model
        response = client.chat.completions.create(
            model=MODEL,
            messages=messages,
            tools=tools,
            temperature=0.2,
            max_tokens=1024
        )
        
        assistant_message = {"role": "assistant"}
        
        # Check if the response includes a tool call
        if hasattr(response.choices[0].message, 'tool_calls') and response.choices[0].message.tool_calls:
            tool_calls = response.choices[0].message.tool_calls
            assistant_message["content"] = response.choices[0].message.content
            assistant_message["tool_calls"] = tool_calls
            messages.append(assistant_message)
            
            print("\n[System: Processing tool calls...]")
            
            # Process each tool call
            for tool_call in tool_calls:
                print(f"[Using {tool_call.function.name}]")
                
                # Parse arguments and call the tool
                args = json.loads(tool_call.function.arguments)
                result = process_tool_call(tool_call.function.name, args)
                
                # Add tool result to messages
                messages.append({
                    "role": "tool",
                    "tool_call_id": tool_call.id,
                    "name": tool_call.function.name,
                    "content": json.dumps(result)
                })
            
            # Get the final response after tool use
            final_response = client.chat.completions.create(
                model=MODEL,
                messages=messages,
                temperature=0.2,
                max_tokens=1024
            )
            
            messages.append({
                "role": "assistant",
                "content": final_response.choices[0].message.content
            })
            
            return final_response.choices[0].message.content
        else:
            # No tool call, just regular response
            assistant_message["content"] = response.choices[0].message.content
            messages.append(assistant_message)
            return response.choices[0].message.content
    
    # Interactive loop - you'll need to run this cell multiple times, once for each message
    user_input = input("User: ")
    
    if user_input.lower() != 'exit':
        response = process_message(user_input)
        print(f"\nTechNova Support: {extract_reply(response)}")
    else:
        print("\nThank you for using TechNova Support!")

```

```python
def run_colored_chatbot():
    """Chatbot with colored terminal output for better readability"""
    try:
        # Check if colorama is available
        from colorama import init, Fore, Style
        init()  # Initialize colorama
        
        # Define colors
        USER_COLOR = Fore.GREEN
        BOT_COLOR = Fore.BLUE
        SYSTEM_COLOR = Fore.YELLOW
        RESET = Style.RESET_ALL
        
    except ImportError:
        # Fallback if colorama is not installed
        print("For colored output, install colorama: pip install colorama")
        USER_COLOR = ""
        BOT_COLOR = ""
        SYSTEM_COLOR = ""
        RESET = ""
    
    client = Together(api_key=os.getenv("TOGETHER_API_KEY"))
    MODEL = "Qwen/Qwen3-235B-A22B-fp8-tput"
    
    system_prompt = """
    You are a customer support chat bot for an online retailer called TechNova.
    Your job is to help users look up their account, orders, and cancel orders.
    Be helpful and brief in your responses.
    You have access to a set of tools, but only use them when needed.
    If you do not have enough information to use a tool correctly, ask a user follow up questions to get the required inputs.
    Do not call any of the tools unless you have the required data from a user.
    
    Place all user-facing conversational responses in <reply></reply> XML tags.
    """
    
    messages = [
        {"role": "system", "content": system_prompt}
    ]
    
    print(f"{SYSTEM_COLOR}=" * 60)
    print("TechNova Customer Support Bot (type 'exit' to end)")
    print("=" * 60, RESET)
    
    while True:
        # Get user input
        user_input = input(f"\n{USER_COLOR}User: {RESET}")
        if user_input.lower() in ['exit', 'quit', 'bye']:
            print(f"\n{BOT_COLOR}Thank you for contacting TechNova Support!{RESET}")
            break
            
        # Add user message
        messages.append({"role": "user", "content": user_input})
        
        # Process conversation possibly through multiple tool calls
        waiting_for_final_response = True
        while waiting_for_final_response:
            # Call the model
            try:
                response = client.chat.completions.create(
                    model=MODEL,
                    messages=messages,
                    tools=tools,
                    temperature=0.3,
                    max_tokens=1024
                )
                
                # Add assistant's response to message history
                assistant_message = {
                    "role": "assistant",
                    "content": response.choices[0].message.content
                }
                
                if hasattr(response.choices[0].message, 'tool_calls') and response.choices[0].message.tool_calls:
                    assistant_message["tool_calls"] = response.choices[0].message.tool_calls
                    
                messages.append(assistant_message)
                
                # Handle tool calls if present
                if hasattr(response.choices[0].message, 'tool_calls') and response.choices[0].message.tool_calls:
                    # Process each tool call
                    for tool_call in response.choices[0].message.tool_calls:
                        try:
                            print(f"\n{SYSTEM_COLOR}[Using tool: {tool_call.function.name}]{RESET}")
                            
                            # Parse and execute tool
                            args = json.loads(tool_call.function.arguments)
                            result = process_tool_call(tool_call.function.name, args)
                            
                            # Format results for better readability
                            if isinstance(result, dict) or isinstance(result, list):
                                formatted_result = json.dumps(result, indent=2)
                                print(f"{SYSTEM_COLOR}Tool result: {formatted_result}{RESET}")
                            else:
                                print(f"{SYSTEM_COLOR}Tool result: {result}{RESET}")
                            
                            # Add tool result to conversation
                            messages.append({
                                "role": "tool",
                                "tool_call_id": tool_call.id,
                                "name": tool_call.function.name,
                                "content": json.dumps(result)
                            })
                        except Exception as e:
                            print(f"{SYSTEM_COLOR}Tool error: {e}{RESET}")
                            messages.append({
                                "role": "tool",
                                "tool_call_id": tool_call.id,
                                "name": tool_call.function.name,
                                "content": f"Error: {str(e)}"
                            })
                else:
                    # No tool calls, show final response
                    waiting_for_final_response = False
                    reply = extract_reply(response.choices[0].message.content)
                    print(f"\n{BOT_COLOR}TechNova Support: {reply}{RESET}")
                    
            except Exception as e:
                print(f"\n{SYSTEM_COLOR}Error: {e}{RESET}")
                waiting_for_final_response = False
    
# Uncomment to run the colored chatbot
# run_colored_chatbot()
```
